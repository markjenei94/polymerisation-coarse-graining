```python
import numpy as np
from tm_develop import TrajectoryMatching
from misc import plot_1component
import time

def basis_function(x, p):
    return np.sign(x) * np.abs(x) ** p

T = 350
outfile = f"/home/markjenei/polymerisation-coarse-graining/dgeba_hardener/epoxy_compass/epoxy_npt_{T}.out"
PATH = "/home/markjenei/polymerisation-coarse-graining/dgeba_hardener/epoxy/"
```

```python
run = 1
steps_between_points = 1
configurations = 1000
params = range(-1, -21, -1)


print(f"T = {T}")
t = time.localtime()
current_time = time.strftime("%H:%M:%S", t)
print("Start at ",current_time)


epoxy_tm = TrajectoryMatching(outfile_path=outfile, basis=basis_function, basis_parameters=params,
                                   simulation_timestep=0.5, cutoff=50, system_style='molecular',
                                   every_n_from_output=steps_between_points, timesteps_in_fit=configurations)

cg_ids_1 = [0,1,2,3,4,5,6,15,16,17,18,23,25,26,27,28,40,41,42,43,36,37,38,39,14]
cg_ids_2 = [6,7,8,9,10,11,12,13,19,20,21,22,24,29,30,31,32,33,34,35,44,45,46,47,48]
for i in range(int(epoxy_tm.data.shape[1] / 49 + 1)):
    epoxy_tm.cg_instructions[i] = [cg_ids_1, cg_ids_2]

epoxy_tm.prepare_training_data()

epoxy_tm.fit()
print("bonded weight = ", epoxy_tm.bonded_weight)

x = np.linspace(10.0, 60, 10000)
y_fit_long = epoxy_tm.predict(x, correct=False)[0]
plot_1component(x, y_fit_long, output_path=f"/home/markjenei/polymerisation-coarse-graining/dgeba_hardener/epoxy/epoxy_npt_{T}_{run}a.png")
x = np.linspace(3.1, 60, 10000)
y_fit_long = epoxy_tm.predict(x, correct=False)[0]
plot_1component(x, y_fit_long, output_path=f"/home/markjenei/polymerisation-coarse-graining/dgeba_hardener/epoxy/epoxy_npt_{T}_{run}b.png")
np.save(f"/home/markjenei/polymerisation-coarse-graining/dgeba_hardener/epoxy/epoxy_npt_{T}_{run}.npy", epoxy_tm.weights)
```
